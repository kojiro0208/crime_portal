```python
# https://di-acc2.com/programming/python/3678/
import requests
import os 
from bs4 import BeautifulSoup

def download_files(url, file_dir, save_name=False):
    os.makedirs(file_dir, exist_ok=True)
    #Web上のファイルデータをダウンロード
    response = requests.get(url)
    
    #HTTP Responseのエラーチェック
    try:
        response_status = response.raise_for_status()
    except Exception as exc:
        print("Error:{}".format(exc))
    
    
    # HTTP Responseが正常な場合は下記実行
    if response_status == None:

        #open()関数にwbを渡し、バイナリ書き込みモードで新規ファイル生成
        if save_name:
            file = open(os.path.join(file_dir, save_name),"wb")
        else:
            file = open(os.path.join(file_dir, os.path.basename(url)),"wb")

        #各チャンクをwrite()関数でローカルファイルに書き込む
        for chunk in response.iter_content(100000):
            file.write(chunk)

        #ファイルを閉じる
        file.close()
        #print("ダウンロード・ファイル保存完了")
```

```python
# 富山県
url_dict = {
    "https://opendata.pref.toyama.jp/dataset/settou2018":2018,
    "https://opendata.pref.toyama.jp/dataset/settou2019":2019,
    "https://opendata.pref.toyama.jp/dataset/settou2020":2020,
    "https://opendata.pref.toyama.jp/dataset/settou2021":2021,
}
file_dir = "./data/toyama/"
base_url = "https://opendata.pref.toyama.jp"

for url, nendo in url_dict.items():
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    li_list = soup.find_all("ul", class_="detail_file")[0].find_all("li")
    data_urls = [l.find("a").get("href") for l in li_list]
    for data_url in data_urls:
        data_url = base_url + data_url
        r2 = requests.get(data_url)
        soup2 = BeautifulSoup(r2.content, "html.parser")
        a_ref = [a.text for a in soup2.find_all("a") if ".csv" in a.text.strip().lower()]
        if len(a_ref) != 0:
            save_name = f"{nendo}_{os.path.basename(a_ref[0])}"
            download_files(a_ref[0], file_dir, save_name)
```

```python
# 千葉
url_dict = {
    "https://www.police.pref.chiba.jp/seisoka/safe-life_publicspace-statistics_00002.html":2021,
    "https://www.police.pref.chiba.jp/seisoka/sample_00050.html":2020,
    "https://www.police.pref.chiba.jp/seisoka/sample_00051.html":2019,
    "https://www.police.pref.chiba.jp/seisoka/sample_00052.html":2018,
}

file_dir = "./data/chiba/"
base_url = "https://www.police.pref.chiba.jp"
for url, nendo in url_dict.items():
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    a_ref = soup.find_all("a")
    csv_urls = []
    csv_urls = [c.get("href") for c in a_ref if ".csv" in c.get("href", default="None")]

    for csv_url in csv_urls:
        download_url = base_url + csv_url
        os.makedirs(file_dir, exist_ok=True)
        response = requests.get(download_url)
        #HTTP Responseのエラーチェック
        try:
            response_status = response.raise_for_status()
        except Exception as exc:
            print("Error:{}".format(exc))

        # HTTP Responseが正常な場合は下記実行
        if response_status == None:
            #open()関数にwbを渡し、バイナリ書き込みモードで新規ファイル生成
            file_name = os.path.basename(download_url).replace(".csv", "")
            file_name = f"{file_name}_{nendo}.csv"
            file = open(os.path.join(file_dir,file_name),"wb")
            #各チャンクをwrite()関数でローカルファイルに書き込む
            for chunk in response.iter_content(100000):
                file.write(chunk)
            #ファイルを閉じる
            file.close()
```
